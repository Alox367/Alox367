```python
import json

nome_file = "1942-04-04_042U0262_VIGENZA_2026-04-22_V0.json"

try:
    with open(nome_file, 'r', encoding='utf-8') as f:
        dati = json.load(f)

    print(f"📦 Struttura principale: {type(dati)}")

    # Se è una lista (come speriamo), analizziamo il primo elemento
    if isinstance(dati, list):
        print(f"📊 Numero totale di elementi: {len(dati)}\n")
        print("🔍 Ecco i 'tag' (chiavi) del primissimo elemento:")

        primo_elemento = dati[0]
        if isinstance(primo_elemento, dict):
            for chiave, valore in primo_elemento.items():
                # Stampiamo la chiave e un'anteprima del suo contenuto (massimo 100 caratteri)
                anteprima = str(valore).replace('\n', ' ')[:100]
                print(f"   ▶ {chiave}: {anteprima}...")
        else:
            print(f"Il primo elemento non è un dizionario, ma: {type(primo_elemento)}")

    # Se invece è un grande dizionario, vediamo le sue sezioni principali
    elif isinstance(dati, dict):
        print("\n🔍 Ecco i 'tag' (chiavi) principali del file:")
        for chiave in dati.keys():
            print(f"   ▶ {chiave}")

except Exception as e:
    print(f"Errore nella lettura: {e}")
```

```python
import json

nome_file = "1942-04-04_042U0262_VIGENZA_2026-04-22_V0.json"

def mappa_albero(dato, livello=0, max_livelli=5):
    spazio = "   " * livello
    # Ci fermiamo se andiamo troppo in profondità per non intasare l'output
    if livello >= max_livelli:
        print(f"{spazio}...")
        return

    if isinstance(dato, dict):
        # Evitiamo di stampare tutto se ci sono troppe chiavi uguali, ne prendiamo massimo 3
        chiavi_da_mostrare = list(dato.keys())[:3]
        for chiave in chiavi_da_mostrare:
            valore = dato[chiave]
            print(f"{spazio}📂 {chiave} ({type(valore).__name__})")
            if isinstance(valore, (dict, list)):
                mappa_albero(valore, livello + 1, max_livelli)
        if len(dato.keys()) > 3:
            print(f"{spazio}  [... e altre {len(dato.keys()) - 3} chiavi simili]")

    elif isinstance(dato, list):
        print(f"{spazio}📑 Lista di {len(dato)} elementi. Il primo è fatto così:")
        if len(dato) > 0:
            mappa_albero(dato[0], livello + 1, max_livelli)

try:
    with open(nome_file, 'r', encoding='utf-8') as f:
        dati = json.load(f)

    print("🔍 Raggi X sull'articolato in corso...\n")
    mappa_albero(dati.get('articolato', {}))

except Exception as e:
    print(f"Errore: {e}")
```

```python
import json

nome_file = "1942-04-04_042U0262_VIGENZA_2026-04-22_V0.json"

try:
    with open(nome_file, 'r', encoding='utf-8') as f:
        dati = json.load(f)

    # Puntiamo dritti al cuore del file usando il percorso che abbiamo scoperto
    primo_elemento = dati.get('articolato', {}).get('elementi', [])[0]

    print("🎯 Bersaglio agganciato! Ecco TUTTE le chiavi del primo elemento:\n")

    for chiave, valore in primo_elemento.items():
        if valore is None:
            print(f"   ▶ {chiave}: Vuoto (None)")
        elif isinstance(valore, str):
            # Se è testo, stampiamo i primi 100 caratteri
            anteprima = valore.replace('\n', ' ')[:100]
            print(f"   ▶ {chiave}: Testo -> '{anteprima}...'")
        elif isinstance(valore, list):
            print(f"   ▶ {chiave}: 📑 Lista con {len(valore)} elementi")
        elif isinstance(valore, dict):
            print(f"   ▶ {chiave}: 📂 Dizionario con {len(valore.keys())} chiavi")
        else:
            print(f"   ▶ {chiave}: Tipo {type(valore).__name__}")

except Exception as e:
    print(f"Errore: {e}")
```

```python
import json

nome_file = "1942-04-04_042U0262_VIGENZA_2026-04-22_V0.json"


def trova_vero_articolo(nodo):
    # Se il nodo è un dizionario, controlliamo le sue chiavi
    if isinstance(nodo, dict):
        tipo = str(nodo.get('tipoNir')).lower()
        numero = str(nodo.get('numNir'))

        # Condizione: Se il tipo è 'articolo' o ha un numero di articolo valido, L'ABBIAMO TROVATO!
        if 'articolo' in tipo or (numero != 'None' and numero != ''):
            return nodo

        # Altrimenti, cerchiamo nelle sue sottocartelle (es. dentro 'elementi')
        for chiave, valore in nodo.items():
            if isinstance(valore, (dict, list)):
                trovato = trova_vero_articolo(valore)
                if trovato: return trovato

    # Se è una lista (es. una lista di Capi o Titoli), esploriamo ogni elemento
    elif isinstance(nodo, list):
        for elemento in nodo:
            trovato = trova_vero_articolo(elemento)
            if trovato: return trovato

    return None

try:
    with open(nome_file, 'r', encoding='utf-8') as f:
        dati = json.load(f)

    print("🕵️‍♂️ Ricerca del primo VERO articolo in corso nelle profondità del file...")

    # Facciamo partire il robot dall'articolato
    articolo_trovato = trova_vero_articolo(dati.get('articolato', {}))

    if articolo_trovato:
        print("\n✅ BINGO! Trovato un articolo. Ecco i suoi dati Pieni:\n")
        for k, v in articolo_trovato.items():
            if v and str(v) != '[]': # Stampiamo SOLO i campi che contengono davvero qualcosa
                anteprima = str(v).replace('\n', ' ')[:150]
                print(f"   ▶ {k}: {anteprima}")
    else:
        print("Nessun articolo trovato. Il formato è più strano del previsto!")

except Exception as e:
    print(f"Errore: {e}")
```
